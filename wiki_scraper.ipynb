```python
from urllib import response
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population"

table_class = 'wikitable sortable jquery-tablesorter'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.find('table',{'class':"wikitable sortable"})
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)

table = soup.find('table',{'class':"wikitable sortable"})
df_raw = pd.read_html(str(table))
df = pd.DataFrame(df_raw[0])
df = df[['City', 'State[c]', '2020census', 'Change', '2020 population density']]
df.rename(columns={"State[c]": "State"}, inplace=True)

df['City'] = df['City'].str.replace(r'\[(.*?)\]', '')
df['2020 population density'] = df['2020 population density'].str.replace(r'\D', '').astype(int)
# df['2020 population density'] = pd.to_numeric(df['2020 population density'])
df = df.sort_values(by='2020 population density', ascending=False).reset_index(drop=True)

df.head()
```

```python
df.to_csv('list_of_cities.csv', index=False)
```
